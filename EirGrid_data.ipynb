```python
import os
import pandas as pd
import matplotlib.pyplot as plt
os.chdir(r"C:\Users\luciano.detommasi\EirGrid_Data\ROI")
```

```python
df_demand_14 = pd.read_csv('ROI_demandactual_14_Eirgrid.csv')
demand_14 = df_demand_14.iloc[:,3].values
plt.title('Year 2014')
plt.xlabel('Time step [15 min.]')
plt.ylabel('Actual Demand [MW]')
plt.plot(demand_14)
plt.show()

df_generation_14 = pd.read_csv('ROI_generationactual_14_Eirgrid.csv')
generation_14 = df_generation_14.iloc[:,3].values
plt.title('Year 2014')
plt.xlabel('Time step [15 min.]')
plt.ylabel('Generation [MW]')
plt.plot(generation_14)
plt.show()

df_interconnection_14 = pd.read_csv('ROI_interconnection_14_Eirgrid.csv')
interconnection_14 = df_interconnection_14.iloc[:,3].values
plt.title('Year 2014')
plt.xlabel('Time step [15 min.]')
plt.ylabel('Interconnection [MW]')
plt.plot(interconnection_14)
plt.show()

df_windactual_14 = pd.read_csv('ROI_windactual_14_Eirgrid.csv')
wind_14 = df_windactual_14.iloc[:,3].values
plt.title('Year 2014')
plt.xlabel('Time step [15 min.]')
plt.ylabel('Wind [MW]')
plt.plot(wind_14)
plt.show()
```
